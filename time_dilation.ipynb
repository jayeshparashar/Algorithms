```python
# Time dialation what is the concept
import numpy as np
```

```python
# set number of ticks to a maximum value if could be largest possible value whatever your computer allows
highest_ticks = 300  

time  = np.arange(highest_ticks+1, 1, -1)
unit_time = 1/time
change_in_space_cor = np.arange(0, highest_ticks, 1)
velocity = change_in_space_cor/unit_time   # velocity is space travelled in unit time
#print(unit_time)
#print(change_in_space_cor)
#print(velocity)
```

```python
import matplotlib.pyplot as plt
# plot time ticks vs. velocity ticks
plt.plot(velocity, time)
plt.xlabel('plot velocity for space ticks per unit time')
plt.ylabel('time ticks ')
plt.title('velocity time plot')
plt.legend(["space-time ticks"])
plt.show()
```

```python
# here in above plot we can see that the 'Changes in space' ( that is  velocity) is zero when
# time ticks at top rate of 300 ticks, however when we adjust counts to increase the changes 
# (number of ticks) in SPACE while decreasing TIME ticks, velocity increaes to a certain value
# and it start decreasing once the unit TIME component in denominator get bigger and takes over
```

```python
# In relativity the ticks in TIME and SPACE doesn't have a linear relation and they have two 
# constraints, #1 both the changes in TIME and SPACE are orthogonal and second the maximum ticks
# for any of two properties is controlled by the speed of light.

# given these constraints the effect of unit time becoming larger kicks in later when velocity 
# v of an object gets closer to speed of light, this is why we ignored smaller v and have 
# started  with v = 0.5*c

c=3
vstart = c/2; vend = c - 0.00001
interval = 0.005
velocity_object = np.arange(vstart, vend, interval)
time_dilation_gamma = 1/np.sqrt(1 - ((velocity_object**2)/(c**2)))
print(time_dilation_gamma)
```

```python
import matplotlib.pyplot as plt_td
# plot time ticks vs. velocity ticks
plt_td.plot(velocity_object, time_dilation_gamma)
plt_td.xlabel('velocity of object')
plt_td.ylabel('time_dilation_gamma')
plt_td.title('velocity time dilation plot')
plt_td.legend(["time dilation"])
plt_td.show()
```
