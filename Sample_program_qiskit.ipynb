```python
# Sample program to get statevector (using statevector simulator) and random distribution 
# (using qasm simulator)) states obtained for a equal superpostion state implemented in qiskit
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, BasicAer, execute, Aer
from qiskit.visualization import plot_histogram, plot_bloch_multivector
```

```python
base = 2
qr = QuantumRegister(base)
qc = QuantumCircuit(qr)
qc.h(range(base))  # put all input qubits in equal superposition by applying H on all lines 
qc.draw()
```

```python
# Run and print results with statevector simulator
bkend = BasicAer.get_backend('statevector_simulator')
shots = 1
job_qasm = execute(qc, bkend, shots=shots)
sv  = job_qasm.result().get_statevector(qc)
print(sv)
```

```python
# plot states on bloch vector
plot_bloch_multivector(sv)
```

```python
# add measurement as we want to run this circuit now on  qasm simulator
qc.measure_all()
qc.draw()
```

```python
# run program on qasm simulator and print results with increasing number of times (shots) 
bkend = BasicAer.get_backend('qasm_simulator')
shots = 1
job_qasm = execute(qc, bkend, shots=shots)
counts  = job_qasm.result().get_counts(qc)
plot_histogram(counts)
```

```python
shots = 10
job_qasm = execute(qc, bkend, shots=shots)
counts  = job_qasm.result().get_counts(qc)
plot_histogram(counts)
```
