```python
# Sample program to get random distribution in qiskit
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, BasicAer, execute
from qiskit.visualization import plot_histogram 
import numpy as np
#import matplotlib.pyplot as plt
```

```python
base = 2
qr = QuantumRegister(base)
qc = QuantumCircuit(qr)
qc.h(range(base))  # put all input qubits in equal superposition by applying H on all lines 
qc.measure_all()
qc.draw()
```

```python
bkend = BasicAer.get_backend('qasm_simulator')
shots = 1
job_qasm = execute(qc, bkend, shots=shots)
counts  = job_qasm.result().get_counts(qc)
plot_histogram(counts)
```
